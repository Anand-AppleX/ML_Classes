```python
X = dataset.iloc[:,:-1]
# X=dataset.drop(dataset['Price'])
X
```

```python
# y = dataset.iloc[:-1]
y = dataset[['Price']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
y_train
```

# Standardize/feature scaling the dataset for feature engineering

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

# model building

```python
from sklearn.linear_model import LinearRegression
```

```python
l_model = LinearRegression()
l_model
```

```python
l_model.fit(x_train,y_train)
```

```python
# Print coefficients and intercept
print(l_model.coef_)
```

```python
print(l_model.intercept_)
```

# prediction

```python
l_pred = l_model.predict(x_test)
```

```python
l_pred
```

```python
plt.scatter(y_test,l_pred)
```

```python
# performance metrics
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_absolute_error(y_test,l_pred))
```

```python
from sklearn.metrics import r2_score
score =r2_score(y_test,l_pred)
print(score)
```
